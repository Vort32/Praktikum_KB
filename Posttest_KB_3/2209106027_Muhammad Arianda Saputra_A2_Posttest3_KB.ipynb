```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AQI and lat Long of Countries.csv")
```

****Scatter Plot****

##### mengkategorikan kualitas udara di seluruh dunia menggunakan scatter plot, bisa diliat dibawah negara negara dengan kualitas udara yang berbeda ditampilkan dengan titik yang berbeda. ada yang negaranya bersih dari polusi dan ada juga negara yang polusinya sangat berbahaya bahkan beracun.

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=df,x='lng',y='lat',hue='AQI Category',palette='Paired')
plt.title('Kategori Indeks Kualitas Udara di seluruh dunia',fontsize=20)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

***PLOT BAR***

##### menampilkan data 5 negara dengan jumlah tertinggi di kolom "negara pada dataset dengan Bar plot. data di ambil sesuai kota yang terdapat pada negara tersebut, bisa diliat di amerika lebih banyak karna data yang diambil banyak di kota amerika.

```python
df=df.Country.value_counts().sort_values(ascending=False)[:5].reset_index()
plt.figure(figsize=(16, 8))
sns.barplot(data=df, y='index', x='Country')
plt.title('Top 5 Nilai Negara paling banyak di dalam Data');
plt.xlabel('Jumlah')
plt.ylabel('Negara')
plt.show()
```

##### menampilkan distribusi kategori indeks kualitas udara dengan plot bar. bisa dilihat dibawah, kualitas udara yang bagus tergolong cukup banyak.

```python
sns.countplot(data=df, x='AQI Category')
plt.title('Distribution of AQI Categories')
plt.xlabel('kategori kualitas udara')
plt.ylabel('perhitungan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

****PIE CHART****

##### menampilkan distribusi kategori indeks kualitas udara dengan diagram pie. diagram ini menggambarkan persentase nilai kategori AQI pada beberapa negara dengan kualitas udara di tiap kotanya. bisa diliat diagram di bawah, amerika lebih dominan daripada negara lain dikarenakan kota kotanya banyak yang di cek kualitas udaranya.  

```python
ratarata = df.groupby('Country').sum().sort_values(by=['AQI Value'],ascending = False).head(5)
```

```python
plt.pie(data=ratarata,x='AQI Value',labels=ratarata.index,autopct='%1.2f%%',shadow=True,explode=(0.1,0,0,0,0))
plt.title('Distribusi nilai kategori indeks kualitas udara di negara-negara dengan pencemaran udara paling banyak di kotanya')
plt.show()
```

****HEAT MAP****

##### menampilkan korelasi indeks kualitas udara dengan heat Map. Korelasi ini mengindikasikan sejauh mana perubahan dalam satu parameter dapat memengaruhi perubahan dalam parameter lainnya. seperti jika dua komponen memiliki korelasi positif yang kuat (nilai mendekati 1 pada heatmap), ini berarti ketika satu komponen meningkat, komponen lainnya cenderung juga meningkat, dan sebaliknya. dengan heat map Ini dapat berguna dalam pemantauan kualitas udara, pemahaman dampak pencemaran udara terhadap kesehatan, dan pengambilan keputusan terkait dengan lingkungan.

```python
data = df[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, cmap='Paired', vmin=-1, vmax=1)
plt.title('korelasi Indeks Kualitas Udara')
plt.show()
```
