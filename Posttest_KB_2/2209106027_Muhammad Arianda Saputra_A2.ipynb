### Untuk Import library pandas serta dideklarasikan sebagai pd

```python
import pandas as pd
```

### untuk membuka atau membaca file CSV

```python
df = pd.read_csv("AQI and lat Long of Countries.csv")
```

### Untuk Menampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka

```python
df.describe()
```

### untuk menampilkan 10 record pertama bertipe angka

```python
df.select_dtypes(include=['number']).head(10)
```

### Menampilkan Hasil Dari Dictionary Informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka

```python
# Deklarasikan
dict = {
    'Jumlah Record  ' : len(df),
    'Jumlah Atribut ': len(df.columns),
    'Nama Atribut   ' : df.columns.tolist(),
    'Jumlah Atribut dengan Tipe Angka ': len(df.select_dtypes(['int64', 'float']).columns)
}

# Menampilkan Hasil
print("Informasi Dataset")
for key, value in dict.items():
    print(f"- {key}: {value}")
```

# Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.

### tampilkan jumlah label dari semua attribute yang bernilai object

```python
atribut_objek = df.select_dtypes(include=['object'])
jumlah_label = atribut_objek.apply(lambda x: x.value_counts().count())

print(jumlah_label)
```
